```python
################import 구간
import speech_recognition as sr #speech_recognition download
import librosa #librosa download
from pydub import AudioSegment #AudioSegment download
import os
from glob import iglob
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sound_kor_to_eng(): #####전체 함수
    sound_to_str, kor_uni = [], [] #변수 선언
    
    for file in iglob('./**.mp3', recursive=True): #####mp3 to wav
        mp3_file = file.replace('.\\','')
        sound = AudioSegment.from_mp3(mp3_file)
        sound.export(mp3_file.replace('.mp3','')+'.wav', format="wav")
    
    for file in iglob('./**.wav', recursive=True): #####wav to sound_text
        wav_file = file.replace('.\\','')

        r=sr.Recognizer()
        sample_wav, rate = librosa.core.load(wav_file)
        korean_audio = sr.AudioFile(wav_file)

        with korean_audio as source:
            audio = r.record(source)
        a_kor = r.recognize_google(audio_data=audio, language='ko-KR')
        sound_to_str.append(a_kor)
        
    ###############sound_text to alphabet
    
    #초성 리스트(분리용)
    CHOSUNG_LIST = ['ㄱ', 'ㄲ', 'ㄴ', 'ㄷ', 'ㄸ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅃ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅉ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
    #중성 리스트(분리용)
    JUNGSUNG_LIST = ['ㅏ', 'ㅐ', 'ㅑ', 'ㅒ', 'ㅓ', 'ㅔ', 'ㅕ', 'ㅖ', 'ㅗ', 'ㅘ', 'ㅙ', 'ㅚ', 'ㅛ', 'ㅜ', 'ㅝ', 'ㅞ', 'ㅟ', 'ㅠ', 'ㅡ', 'ㅢ', 'ㅣ']
    #종성 리스트(분리용)
    JONGSUNG_LIST = [' ', 'ㄱ', 'ㄲ', 'ㄳ', 'ㄴ', 'ㄵ', 'ㄶ', 'ㄷ', 'ㄹ', 'ㄺ', 'ㄻ', 'ㄼ', 'ㄽ', 'ㄾ', 'ㄿ', 'ㅀ', 'ㅁ', 'ㅂ', 'ㅄ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
    
    #초성 딕셔너리(발음용)
    CHOSUNG_DICT = {'ㄱ': ['g','g','g'], 'ㄴ':['n','n','n'],'ㄷ':['d','d','d'],'ㄹ':['r','r','r'],'ㅁ':['m','m','m'],
                    'ㅂ':['b','b','b'],'ㅅ':['s','s','s'],'ㅇ':['','',''],
                    'ㅈ':['j','j','j'],'ㅊ':['ch','ch','ch'],'ㅋ':['k','k','k'],'ㅌ':['t','t','t'], 'ㅍ':['p','p','p'],'ㅎ':['h','h','h'],
               'ㄲ':['k','k','k'],'ㄸ':['t','t','t'],'ㅃ':['p','p','p'],'ㅆ':['s','s','s'],'ㅉ':['j','j','j'],' ': ['-','-','-']}
    #종성 딕셔너리(발음용)
    JONGSUNG_DICT = {'ㄱ': ['k','k','k'], 'ㄴ':['n','n','n'],'ㄷ':['t','t','t'],'ㄹ':['l','l','l'],'ㅁ':['m','m','m'],'ㅂ':['p','p','p'],
                     'ㅅ':['t','t','t'],'ㅇ':['ng','ng','ng'],'ㅈ':['j','j','j'],'ㅊ':['t','t','t'],'ㅋ':['k','k','k'],
                     'ㅌ':['t','t','t'], 'ㅍ':['p','p','p'],'ㅎ':['t','t','t'],'ㄲ':['k','k','k'],'ㄸ':['t','t','t'],
                     'ㅃ':['p','p','p'],'ㅆ':['t','t','t'],'ㅉ':['t','t','t'], 'ㄳ':['k','k','k'],'ㄵ':['n','n','n'],
                     'ㅄ':['p','p','p'],'ㄶ':['n','n','n'],'ㅀ':['l','l','l'],
                     'ㄼ':['l','l','l'],'ㄽ':['l','l','l'],'ㄾ':['l','l','l'],'ㄺ':['k','k','k'],'ㄻ':['m','m','m'],
                     'ㄿ':['p','p','p'],' ': ['-','-','-']}
    #중성 딕셔너리(발음용)
    JUNGSUNG_DICT={'ㅏ':['a','a','a'],'ㅐ':['a','e','e'],'ㅑ':['y','a','a'],'ㅒ':['y','a','e'],'ㅓ':['e','o','o'],'ㅔ':['e','e','e'],
                   'ㅖ':['y','e','e'],'ㅕ':['y','e','o'],'ㅗ':['o','o','o'],'ㅘ':['w','a','a'],
                  'ㅙ':['w','a','e'],'ㅚ':['o','e','e'],'ㅛ':['y','o','o'],'ㅜ':['oo','oo','oo'],'ㅝ':['w','o','o'],'ㅞ':['w','e','e'],
                   'ㅟ':['w','i','i'],'ㅠ':['y','u','u'],'ㅡ':['e','u','u'],'ㅣ':['i','i','i'],'ㅢ':['u','i','i']}
    
    #####sound_text to kor_alphabet
    for sound_to_str_num in range(len(sound_to_str)):
        def korean_to_be_englished(korean_word):
            r_lst = []
            for w in list(korean_word.strip()):
                if '가'<=w<='힣':
                    ch1 = (ord(w) - ord('가'))//588
                    ch2 = ((ord(w) - ord('가')) - (588*ch1)) // 28
                    ch3 = (ord(w) - ord('가')) - (588*ch1) - 28*ch2
                    if JONGSUNG_LIST[ch3] == ' ':
                        #print([CHOSUNG_LIST[ch1], JUNGSUNG_LIST[ch2]])
                        r_lst.append([CHOSUNG_LIST[ch1], JUNGSUNG_LIST[ch2]])
                    else:
                        #print([CHOSUNG_LIST[ch1], JUNGSUNG_LIST[ch2], JONGSUNG_LIST[ch3]])
                        r_lst.append([CHOSUNG_LIST[ch1], JUNGSUNG_LIST[ch2], JONGSUNG_LIST[ch3]])
                else:
                    r_lst.append([w])
            return r_lst

        kor_uni.append(korean_to_be_englished(sound_to_str[sound_to_str_num]))
        
    #####kor_alphabet to eng_alphabet
    trans_kor_to_eng_sub = [[] for _ in range(len(kor_uni))]
    trans_kor_to_eng = [[] for _ in range(len(kor_uni))]
    for j in range(len(kor_uni)):
        for i in range(len(kor_uni[j])):
            for k in range(len(kor_uni[j][i])):
                if k == 0 and CHOSUNG_DICT[kor_uni[j][i][k]] == ['','','']:
                    continue;
                elif k == 0:
                    for kk1 in range(len(CHOSUNG_DICT[kor_uni[j][i][k]])):
                        trans_kor_to_eng_sub[j].append(CHOSUNG_DICT[kor_uni[j][i][k]][kk1])
                elif k == 1:
                    for kk2 in range(len(JUNGSUNG_DICT[kor_uni[j][i][k]])):
                        trans_kor_to_eng_sub[j].append(JUNGSUNG_DICT[kor_uni[j][i][k]][kk2])
                elif k == 2: 
                    for kk3 in range(len(JONGSUNG_DICT[kor_uni[j][i][k]])):
                        trans_kor_to_eng_sub[j].append(JONGSUNG_DICT[kor_uni[j][i][k]][kk3])
        trans_kor_to_eng[j] = ''.join(trans_kor_to_eng_sub[j])
        #trans_kor_to_eng_sub.replace('','(term)')
        
    for img_len in range(len(trans_kor_to_eng_sub)):
        img_array = []
        for img_fixel_len in range(len(trans_kor_to_eng_sub[img_len])):
            #print(trans_kor_to_eng_sub[img_len][img_fixel_len])
            img = cv2.imread('./alphabet_img_all_boy_ver/'+trans_kor_to_eng_sub[img_len][img_fixel_len]+'.jpg', cv2.IMREAD_COLOR)
            height, width, layers = img.shape
            size = (width, height)
            img_array.append(img)

        out = cv2.VideoWriter(trans_kor_to_eng_sub[img_len][img_fixel_len]+'.mp4', cv2.VideoWriter_fourcc(*'DIVX'),5,size)

        for avi_count in range(len(img_array)):
            out.write(img_array[avi_count])
        out.release()
    
    #return trans_kor_to_eng, trans_kor_to_eng_sub # total result
    
    
    
sound_kor_to_eng()
```
